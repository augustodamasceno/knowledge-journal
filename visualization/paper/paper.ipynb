# Summary

# Libraries  

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

# Example 1: Line Plot of Two Variables (1 X-axis, 1 Y-axis)

```python
# DATA
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
y1 = np.sin(x)
y2 = np.cos(x)

# Configure the plot
dpi = 150
filename = "example1"
filetype = "png"
title = r"Title"
label1 = r"$y_{1}$"
label2 = r"$y_{2}$"
xlabel = r"$x$"
ylabel = r"$f(x)$"
width = 8
height = 4

linewidth = 1.0
linewidth_axis = 0.8
linewidth_grid_lines_major = 0.7
linewidth_grid_lines_minor = 0.5
font = "DejaVu Sans"
font_size_title = 14
font_size_axis = 12
color1 = "black"
color2 = "gray"
color_axis = "black"
legend_position =  "upper right"
color_grid_line_major = "gray"
color_grid_line_minor = "gray"
grid_lines_major_multiplicator_x = 1.0
grid_lines_major_multiplicator_y = 0.5
grid_lines_minor_multiplicator_x = 0.5
grid_lines_minor_multiplicator_y = 0.25

# Font
plt.rcParams.update({
    "font.family": font,
    "font.size": font_size_title,
    "text.usetex": False,
    "axes.grid": True,
    "grid.linestyle": "--",
    "grid.alpha": 0.5
})



# Create figure
fig, ax = plt.subplots(figsize=(width, height), dpi=dpi)

# Plot sine and cosine
ax.plot(x, y1, label=label1, color=color1, linewidth=linewidth)
ax.plot(x, y2, label=label2, color=color1, linestyle="dashed", linewidth=linewidth)

# Labels and title
ax.set_title(title, fontsize=font_size_title)
ax.set_xlabel(xlabel, fontsize=font_size_axis)
ax.set_ylabel(ylabel, fontsize=font_size_axis)

# Axes lines
ax.axhline(0, color=color_axis, linewidth=linewidth_axis)
ax.axvline(0, color=color_axis, linewidth=linewidth_axis)

# Enable grid
plt.grid(True)

# Grid Lines
plt.grid(which='major', color=color_grid_line_major, linestyle='-', linewidth=linewidth_grid_lines_major)
plt.grid(which='minor', color=color_grid_line_minor, linestyle=':', linewidth=linewidth_grid_lines_minor)

# Set grid spacing (e.g., every 1 unit for x and y axes)
plt.gca().xaxis.set_major_locator(MultipleLocator(grid_lines_major_multiplicator_x))  # Major ticks on x every 1 unit
plt.gca().yaxis.set_major_locator(MultipleLocator(grid_lines_major_multiplicator_y))  # Major ticks on y every 0.2 units

# Optional: Enable minor ticks and grid lines for finer control
plt.minorticks_on()
plt.gca().xaxis.set_minor_locator(MultipleLocator(grid_lines_minor_multiplicator_x))  # Minor ticks on x every 0.5 units
plt.gca().yaxis.set_minor_locator(MultipleLocator(grid_lines_minor_multiplicator_y))  # Minor ticks on y every 0.1 units

# Legend
ax.legend(frameon=False, loc=legend_position)

# Save as high-resolution figure
plt.tight_layout()
plt.savefig(f"{filename}.{filetype}", dpi=dpi, bbox_inches="tight")

# Show plot
plt.show()
```

# Example 2: Line Plot of One Variable  

```python
# DATA
x = np.linspace(-0.5 * np.pi, 0.5 * np.pi, 400)
y = np.atan(x)

# Configure the plot
dpi = 150
filename = "example2"
filetype = "png"
title = r"Title"
label1 = r"$y_{1}$"
xlabel = r"$x$"
ylabel = r"$f(x)$"
width = 8
height = 4

linewidth = 1.0
linewidth_axis = 0.8
linewidth_grid_lines_major = 0.7
linewidth_grid_lines_minor = 0.5
font = "DejaVu Sans"
font_size_title = 14
font_size_axis = 12
color1 = "black"
color_axis = "black"
legend_position =  "upper right"
color_grid_line_major = "gray"
color_grid_line_minor = "gray"
grid_lines_major_multiplicator_x = 0.5
grid_lines_major_multiplicator_y = 0.25
grid_lines_minor_multiplicator_x = 0.1
grid_lines_minor_multiplicator_y = 0.1

# Font
plt.rcParams.update({
    "font.family": font,
    "font.size": font_size_title,
    "text.usetex": False,
    "axes.grid": True,
    "grid.linestyle": "--",
    "grid.alpha": 0.5
})


# Create figure
fig, ax = plt.subplots(figsize=(width, height), dpi=dpi)

# Plot sine and cosine
ax.plot(x, y, label=label1, color=color1, linewidth=linewidth)

# Labels and title
ax.set_title(title, fontsize=font_size_title)
ax.set_xlabel(xlabel, fontsize=font_size_axis)
ax.set_ylabel(ylabel, fontsize=font_size_axis)

# Axes lines
ax.axhline(0, color=color_axis, linewidth=linewidth_axis)

# Enable grid
plt.grid(True)

# Grid Lines
plt.grid(which='major', color=color_grid_line_major, linestyle='-', linewidth=linewidth_grid_lines_major)
plt.grid(which='minor', color=color_grid_line_minor, linestyle=':', linewidth=linewidth_grid_lines_minor)

# Set grid spacing (e.g., every 1 unit for x and y axes)
plt.gca().xaxis.set_major_locator(MultipleLocator(grid_lines_major_multiplicator_x))  # Major ticks on x every 1 unit
plt.gca().yaxis.set_major_locator(MultipleLocator(grid_lines_major_multiplicator_y))  # Major ticks on y every 0.2 units

# Optional: Enable minor ticks and grid lines for finer control
plt.minorticks_on()
plt.gca().xaxis.set_minor_locator(MultipleLocator(grid_lines_minor_multiplicator_x))  # Minor ticks on x every 0.5 units
plt.gca().yaxis.set_minor_locator(MultipleLocator(grid_lines_minor_multiplicator_y))  # Minor ticks on y every 0.1 units

# Save as high-resolution figure
plt.tight_layout()
plt.savefig(f"{filename}.{filetype}", dpi=dpi, bbox_inches="tight")

# Show plot
plt.show()
```
